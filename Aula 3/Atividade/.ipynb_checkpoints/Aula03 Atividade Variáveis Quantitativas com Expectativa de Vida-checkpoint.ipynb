```python
dados = pd.read_excel('Life.xlsx')
```

```python
dados.head()
```

```python
dados_1800 = dados.L1800
dados_1800.value_counts()
```

```python
dados_1926 = dados.L1926
dados_1926.value_counts()
```

```python
dados_2000 = dados.L2000
dados_2000.value_counts()
```

```python
dados_1800.describe()
```

```python
dados_1926.describe()
```

```python
dados_2000.describe()
```

```python
from numpy import arange

minimo = dados_1800.min() // 1
maximo = (dados_2000.max() // 1)

intervalo = (maximo - minimo)**(1/2)
print(intervalo)
intervalo = 8

faixa = arange(minimo, maximo, intervalo)
print("Faixas = ", faixa)
```

```python
dados_1800_cat = pd.cut(dados_1800, faixa, False)
dados_1926_cat = pd.cut(dados_1926, faixa, False)
dados_2000_cat = pd.cut(dados_2000, faixa, False)
```

```python
pd.DataFrame([dados_1800,dados_1800_cat]).T
```

```python
pd.DataFrame([dados_1926,dados_1926_cat]).T
```

```python
pd.DataFrame([dados_2000,dados_2000_cat]).T
```
